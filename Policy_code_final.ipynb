# Feature selection 

```python
# Train_test_split

features=df4.columns.difference(['Response'])

from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(df4[features],
                                              df4['Response'],
                                              test_size=0.25,
                                              random_state=43)
```

```python
# Run the base model

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn import metrics
from sklearn import ensemble
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,recall_score,accuracy_score,precision_score
```

```python
seed=0
dt=DecisionTreeClassifier(random_state=seed)
rf= RandomForestClassifier(random_state=seed)
lr= LogisticRegression(random_state= seed)
adb= ensemble.AdaBoostClassifier()
bgc=ensemble.BaggingClassifier()
gbc=ensemble.GradientBoostingClassifier()
xgb=XGBClassifier(random_state=seed)
svc=SVC(random_state=seed)

models=[dt,rf,lr,adb,bgc,gbc,svc,xgb]

def score_model(train_x,train_y,test_x, test_y):
    g_columns=[]
    g=pd.DataFrame(columns=g_columns)
    i=0
    
    for model in models:
        model.fit(train_x,train_y)
        y_pred=model.predict(test_x)
        
        train_accuracy=model.score(train_x,train_y)
        test_accuracy=model.score(test_x,test_y)
        
        p_score= metrics.precision_score(test_y,y_pred)
        r_score=metrics.recall_score(test_y,y_pred)
        f1_score=metrics.f1_score(test_y,y_pred)
        fp, tp, th= metrics.roc_curve(test_y,y_pred)
        auc = metrics.auc(fp, tp)
        
        g.loc[i,"Model_Name"]=model.__class__.__name__
        g.loc[i,"Precision"]=round(p_score, 2)
        g.loc[i,"Recall"] = round(r_score,2)
        g.loc[i,"Train_Accuracy"]=round(train_accuracy,2)
        g.loc[i,"Test_Accuracy"]=round(test_accuracy,2)
        g.loc[i,"F1_score"]=round(f1_score,2)
        g.loc[i,'Auc']=metrics.auc(fp,tp)
        
        i+=1
        
    g.sort_values(by=['F1_score'],ascending=False,inplace=True)
    return(g)
        
```

```python
report = score_model(train_x,train_y,test_x,test_y)
report
```

# Ensemble of Ensemble

Now we will use three different models to put into our Voting Classifier.

```python
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier()
params_rf = {'n_estimators':[50, 100, 200]}
rf_gs = GridSearchCV(rf, params_rf, cv=3)
rf_gs.fit(train_x, train_y)
```

```python
rf_best = rf_gs.best_estimator_
print(rf_gs.best_params_)
```

```python
xgbc=XGBClassifier()
params_xgbc = {'n_estimators':[50, 100, 200]}
xgbc_gs = GridSearchCV(xgbc, params_xgbc, cv=3)
xgbc_gs.fit(train_x, train_y)

xgbc_best = xgbc_gs.best_estimator_
print(xgbc_gs.best_params_)
```

```python
log_reg = LogisticRegression()
log_reg.fit(train_x, train_y)
```

Now we place three models in an array called 'estimators'. Next we will create our voting classifier. It takes two inputs, the first is our estimator array of our three models. we will set the voting parameter to hard, which tells our classifier to make prediciton by majority vote.

```python
from sklearn.ensemble import VotingClassifier
estimators = [('rf', rf_best), ('xgbc', xgbc_best), ('log_reg', log_reg)]
ensemble = VotingClassifier(estimators, voting='hard')
ensemble.fit(train_x, train_y)
ensemble.score(test_x, test_y)
```
